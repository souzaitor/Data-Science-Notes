***
Uma possibilidade é mudarmos o estilo da plotagem - ao invés de termos linhas ligando cada ponto e formando nosso gráfico contínuo, usaremos o parâmetro ```ls``` = (de "line style", "estilo de linha" em inglês") recebendo o valor dotted ("pontilhado").

````python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'g', ls = 'dotted')

eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['Temperatura'], loc = 'lower right', fontsize=15)
````

````python
import datetime
````

***
Seria interessante se conseguíssemos enfatizar determinados períodos de tempo, por exemplo entre 2014 e 2015. Para isso, removeremos o estilo de linha adicionado anteriormente e usaremos a instrução ```eixo.set_xlim()``` para definirmos um limite no eixo X. Essa função recebe dois valores: o primeiro delimitando o limite inicial, e o segundo delimitando o limite final. 

Usaremos a função datetime.datime() para passarmos ambos os valores - o primeiro, 2014,1,1; e o segundo 2015,1,1. Esta função, por padrão, recebe os valores na ordem "ano, mês, dia", portanto estamos delimitando o período entre 1° de janeiro de 2014 e 1° de janeiro de 2015.

````python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'g')

eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
````

Note que o início e o final do ano são períodos mais frios, e a metade do ano apresenta temperaturas mais quentes. Isso nos permite inferir que estamos trabalhando com dados do hemisfério norte. Em países como o Brasil, por exemplo, esse gráfico estaria invertido

***
Continuando, passaremos para a adição de marcadores à nossa visualização. Na função ```plot()```, incluiremos o parâmetro marker (de "marcador" em inglês) recebendo o parâmetro o. Isso fará com que os pontos do nosso gráfico sejam marcados com bolinhas.

````python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'g', marker = 'o')

eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
````

***
Uma customização que pode nos ser interessante é a adição de uma grade ("grid" no inglês) que nos desse uma impressão de continuidade, algo que faz sentido quando pensamos no decorrer de um ano. Para isso, usaremos a instrução ```eixo.grid()``` recebendo o valor True

````python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'g')

eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
eixo.grid(True)
````

***
Podemos fazer uma visualização dentro de outra visualização, na mesma figura, como por exemplo uma visualização com foco no ano de 2014 e em um dos cantos da figura toda a série histórica

Primeiro declaramos a figura com o seu tamanho, criaremos a variável fig usando o comando plt.figure() que já usamos nas outras aulas e o figsize se mantém o mesmo, com o valor de 15,8. Agora declaramos os eixos com uma lista de onde começa e onde termina o eixo. O primeiro eixo a ser definido é da figura maior, contendo os valores 0,0,1,1 e o eixo2 utilizado é para criarmos a figura menor, com os valores 0.1,0.5,0.4,0.4.

````python
fig = plt.figure(figsize=(15,8))

eixo = fig.add_axes([0, 0, 1, 1])
eixo2 = fig.add_axes([0.1, 0.5, 0.4, 0.4])
````

***
No eixo maior vamos colocar a série toda de dados, para fazer isso usamos a função ```eixo.plot()``` passando como argumentos df['data'] e df['temperatura'], representando,respectivamente, os eixos Y e X, fazendo como já fizemos antes, e utilizamos o parâmetro color = 'g', para manter a cor verde. Também vamos adicionar, antes da linha de definição do que é o eixo maior, o ```grid()``` no gráfico e passamos o parâmetro True.

Depois disso isso, usaremos novamente a função ```eixo.set_title()``` para atribuirmos um título ao gráfico. Para evitarmos ambiguidades, passaremos a chamá-lo de "Temperatura no momento", definimos o parâmetro fontsize com valor de 25 e para deixar o título mais acima do gráfico, afim de deixar mais clara a visualização, usamos o parâmetro pad = 20. Por fim, colocamos uma legenda usando a função ```eixo.legend(```) e passamos uma lista,que da nome a característica que estamos observando, nesse caso 'temperatura'. Também definimos mais dois parâmetros nessa função, a lozalização usando loc e passamos 'best' como valor, para que o próprio matplotlib decidda onde é o melhor lugar para colocar a legenda, e mantemos o fontsize com valor de 15.

````python
fig = plt.figure(figsize=(15,8))

eixo = fig.add_axes([0, 0, 1, 1])
eixo2 = fig.add_axes([0.1, 0.5, 0.4, 0.4])

eixo.grid(True)
eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_title('Temperatura no momento', fontsize=25, pad=20)
eixo.legend(['temperatura'], loc='best', fontsize=15)
````

***
Usando a função ```eixo.plot(```) passamos os parâmetros semelhantes ao eixo maior, porém, mudamos a cor da visualização, dessa vez colocamos color = 'b' para que o gráfico seja azul. Antes de continuar a definição do eixo menor, complementamos a definição do eixo que já tinha sido feito, como faltavam os nomes dos eixos X e Y na imagem, vamos definir primeiro o eixo Y com a função ```eixo.set_ylabel()```, passando o nome 'Temperatura' e o fontsize com o valor de 20, para o eixo X, fazemos um trabalho semelhante, mas utilizando a função ```eixo.set_xlabel()``` e o nome que será mostrado no eixo é 'Data', o fontsize continua com o valor de 20.

Depois de fazermos isso, copiamos essas linhas de definições para o eixo menor, ou seja, as linhas que contrém o título, legenda e os nomes dos eixos X e Y, mas precisamos mudar de eixo para eixo2, para que essas definições sejam alocadas no eixo menor da imagem. Para adequar o tamanho dos textos com o tamanho do eixo, vamos alterar alguns valores de fontsize. No título, deixamos o fontsize com valor de 15 e retiramos o parâmetro pad, já que não existe problema do título estar mais próximo do gráfico. Na legenda usamos o valor 8 para o fontsize e finalmente para os eixos Xe Y, definimos fontsize com valor de 10.

````python
fig = plt.figure(figsize=(15,8))

eixo = fig.add_axes([0, 0, 1, 1])
eixo2 = fig.add_axes([0.1, 0.5, 0.4, 0.4])

eixo.grid(True)
eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_title('Temperatura no momento', fontsize=25, pad=20)
eixo.legend(['temperatura'], loc='best', fontsize=15)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)

eixo2.plot(df['data'], df['temperatura'], color = 'b')
eixo2.set_title('Temperatura no momento', fontsize=15)
eixo2.legend(['temperatura'], loc='best', fontsize=8)
eixo2.set_ylabel('Temperatura', fontsize=10)
eixo2.set_xlabel('Data', fontsize=10)
````

***
Outro ajuste interessante seria realçarmos em verde, na visualização menor, as temperaturas do mês de maio, gerando uma associação cromática entre os dois gráficos. A ideia é aplicarmos um filtro que corresponda somente à área desejada do gráfico

Na prática, substituiremos a cor original da visualização menor (eixo2) para g, ou seja, verde. Em seguida, criaremos dois filtros: azul_esquerda, que corresponde ao período anterior ao mês de maio; e azul_direita, correspondendo ao período posterior a tal mês. Usaremos esses filtros para sobrepor, com a cor azul, os períodos selecionados.

Para obtermos esses valores, usaremos ```df['data'] < datetime.datetime(2014,5,1)```, pegando todos os valores da coluna data anteriores ao primeiro dia de maio, e ```df['data'] > datetime.datetime(2014,6,1)``` pegando os valores posteriores ao primeiro dia de junho.

Para aplicarmos nossos filtros, repetiremos a estrutura da função ```eixo2.plot()```. Assim como usávamos df['data'] e df['temperatura'], faremos uma nova plotagem usando ```df[azul_esquerda]['data']``` e ```df[azul_esquerda]['temperatura']```, em seguida repetindo o processo para azul_direita.

````python
fig = plt.figure(figsize=(15,8))

eixo = fig.add_axes([0, 0, 1, 1])
eixo2 = fig.add_axes([0.7, 0.65, 0.3, 0.3])

eixo.grid(True)
eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_xlim(datetime.datetime(2014,5,1),datetime.datetime(2014,6,1))
eixo.set_ylim(270, 320)
eixo.set_title('Temperatura em maio/2014', fontsize=25, pad=20)
eixo.legend(['temperatura'], loc='lower right', fontsize=15)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)

azul_esquerda = df['data'] < datetime.datetime(2014,5,1)
azul_direita = df['data'] > datetime.datetime(2014,6,1)

eixo2.plot(df['data'], df['temperatura'], color = 'g')
eixo2.plot(df[azul_esquerda]['data'],df[azul_esquerda]['temperatura'], color = 'b')
eixo2.plot(df[azul_direita]['data'],df[azul_direita]['temperatura'], color = 'b')
eixo2.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo2.set_title('Temperatura em 2014', fontsize=15)
eixo2.legend(['temperatura'], loc='best', fontsize=8)
eixo2.set_ylabel('Temperatura', fontsize=10)
eixo2.set_xlabel('Data', fontsize=10)
eixo2.grid(True)
````

***
limitaremos o ponto máximo e o ponto mínimo utilizando barras. Para isso, usaremos ```eixo.axhline()```, uma função que nos permitirá gerar uma linha no eixo horizontal (horizontal line). Como parâmetros, passaremos primeiramente ```max()``` - uma função que nos retorna o valor máximo em determinado conjunto de dados - recebendo como parâmetro df['Temperatura']', seguida de color = 'k' para definirmos que nossa linha terá a cor preta, e linestyle = '--', definindo o estilo tracejado para tal linha. 

````python
fig = plt.figure(figsize = (15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_title('Temperatura no momento', fontsize=25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

eixo.axhline(max(df['temperatura']), color = 'k', linestyle='--')
eixo.axhline(min(df['temperatura']), color = 'k', linestyle='--')
````

````python
fig = plt.figure(figsize = (15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_title('Temperatura no momento', fontsize=25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])

eixo.annotate("Máximo", xy=(x1,y1), fontsize=20)

eixo.axhline(max(df['temperatura']), color = 'k', linestyle='--')
eixo.axhline(min(df['temperatura']), color = 'k', linestyle='--')
````

````python
fig = plt.figure(figsize = (15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_title('Temperatura no momento', fontsize=25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmax() - 7000]
y2 = max(df['temperatura']) - 5

eixo.annotate("Máximo", xy=(x1,y1), fontsize=20,
              xytext=(x2,y2), arrowprops=dict(facecolor='k'))

eixo.axhline(max(df['temperatura']), color = 'k', linestyle='--')
eixo.axhline(min(df['temperatura']), color = 'k', linestyle='--')
````

***
Uma alternativa é deslocarmos o texto "Máximo" para uma posição inferior e utilizarmos uma seta para indicar o ponto a que ele se refere.

Primeiramente, sabemos que nossa seta precisará de um ponto final e de um ponto inicial - ou seja, x1,y1 (o ponto de destino da seta, que já temos) e x2,y2. Queremos que o ponto de origem seja próximo, e, pensando nisso, copiaremos e colaremos as definições de x1 e y1, substituindo a cópia pelas novas variáveis. 

Para que o ponto x2 seja plotado à esquerda de x1, subtrairemos o valor do índice (df['temperatura'].idxmax()) por 7000 (equivalente a 7000 horas). Já y2, que se refere ao eixo vertical, reduziremos em 5 (equivalente a 5°K). 

Quanto à anotação em sim, manteremos o texto "Máximo" com o mesmo tamanho de fonte, e acrescentaremos uma posição para o texto com o parâmetro xytext, que receberá os valores x2,y2

Para incluirmos uma seta, usaremos o parâmetro arrowprops que receberá um dicionario (dict()) contendo um argumento facecolor, que definirá a cor da seta. Usaremos k para definir a cor preta.

````python
fig = plt.figure(figsize = (15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_title('Temperatura no momento', fontsize=25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmax() - 7000]
y2 = max(df['temperatura']) - 5

eixo.annotate("Máximo", xy=(x1,y1), fontsize=20,
              xytext=(x2,y2), arrowprops=dict(facecolor='k'))

x1 = df['data'][df['temperatura'].idxmin()]
y1 = min(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmin() - 7000]
y2 = min(df['temperatura']) + 5

eixo.annotate("Mínimo", xy=(x1,y1), fontsize=20,
              xytext=(x2,y2), arrowprops=dict(facecolor='k'))

eixo.axhline(max(df['temperatura']), color = 'k', linestyle='--')
eixo.axhline(min(df['temperatura']), color = 'k', linestyle='--')
````
