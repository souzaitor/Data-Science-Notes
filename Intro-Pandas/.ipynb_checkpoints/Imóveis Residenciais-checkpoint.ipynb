## Imóveis Residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('../csv/aluguel.csv', sep = ';')
dados.head(10)
```

```python
# Lista criada com todos os tipos de imóveis
residencial = list(dados['Tipo'].drop_duplicates())
residencial
```

```python
# Selecionar apenas tipos residenciais
residencial = ['Quitinete', 
'Casa',
'Apartamento',
'Casa de Condomínio',
'Casa de Vila']
```

```python
# Verifica se os elementos são residenciais
dados['Tipo'].isin(residencial).head(10)
```

```python
# Criar DataFrame apenas com imóveis residenciais
selecao = dados['Tipo'].isin(residencial)
dados_residencial = dados[selecao]
dados_residencial
```

```python
# Verificar Integridade dos Dados
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
# Verificar tamanho
dados_residencial.shape[0] 
```

```python
# Construir Index
dados_residencial.index = range(dados_residencial.shape[0])
dados_residencial
```

## Exportando a Base de Dados

```python
# Exportar como CSV
dados_residencial.to_csv('../csv/aluguel_residencial.csv', sep = ';')
```
